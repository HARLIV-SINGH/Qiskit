# Quantum Gates

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_bloch_vector
import matplotlib as mpl
import numpy as np
```

## Hadamard Gate

Creates a superpostion

Used to switch bases ( 0 to '+', 1 to '-', and vice versa)

```python
#Creating a circuit
qcH = QuantumCircuit(1)

#Hadamard gate
qcH.h(0)

#Run on Aer simulator
sim = Aer.get_backend('aer_simulator')
qcH.save_statevector()
state = sim.run(qcH).result().get_statevector(qcH)

#Get output
print(state.data)
```

This is the '+' state, indicating a superposition.

```python
coords = [1,3.14/2,0]  # [Radius, Theta, Phi]
plot_bloch_vector(coords, coord_type='spherical')
```

## Pauli X Gate

A Pauli X gate is equivalent to a bit flip.

You can also think of it as rotating the state about the x
axis by pi on the bloch sphere.

```python
#Create a circuit
qcX = QuantumCircuit(1)

#Pauli X gate (Bit flip)
qcX.x(0) # '0' to '1' (Bit flip)

#Run on Aer sim
sim = Aer.get_backend('aer_simulator')
qcX.save_statevector()
state = sim.run(qcX).result().get_statevector(qcX)

#Get state
print(state.data)
```

This is the '1' state, indicating a bit flip.

```python
coords = [1,3.14,0]  # [Radius, Theta, Phi]
plot_bloch_vector(coords, coord_type='spherical')
```

## Pauli Z gate

A Pauli Z gate is a phase flip.

Equivalent to rotating the state about the z axis by pi on the bloch sphere.

```python
#Create a circuit
qcZ = QuantumCircuit(1)

#Pauli Z gate (Phase flip)
qcZ.h(0) # '0' to '+'
qcZ.z(0) # '+' to '-' (Phase flip)

#Run on Aer sim
sim = Aer.get_backend('aer_simulator')
qcZ.save_statevector()
state = sim.run(qcZ).result().get_statevector(qcZ)

#Get state
print(state.data)
```

This is the '-' state, intdicating a phase flip.

```python
coords = [1,-3.14/2,0]  # [Radius, Theta, Phi]
plot_bloch_vector(coords, coord_type='spherical')
```

## Pauli Y gate

Phase and bit flip

Equivalent to rotating the state about the y axis by pi on the bloch sphere.

```python
#Create a circuit
qcY = QuantumCircuit(1)

#Pauli Y gate (Phase and bit flip)
qcY.h(0) # '0' to '+'
qcY.y(0) # '+' to '−i∣−⟩'

#Run on Aer sim
sim = Aer.get_backend('aer_simulator')
qcY.save_statevector()
state = sim.run(qcY).result().get_statevector(qcY)

#Get state
print(state.data)
```

This is the '−i∣−⟩' state, indicating a rotation about the y axis

```python
coords = [1,3.14/2,-3.14/2]  # [Radius, Theta, Phi]
plot_bloch_vector(coords, coord_type='spherical')
```

## CNOT Gate

Equivalent to an XOR gate


If the control qubit is ∣1⟩, applies an X gate to the target qubit

If the control qubit is ∣0⟩, does nothing

```python
#Create a circuit
qcCX = QuantumCircuit(2)

#Make the control qubit '1'
qcCX.x(0)

#Apply the CNOT gate
qcCX.cx(0, 1)

#Run on Aer simulator
sim = Aer.get_backend('aer_simulator')
qcCX.save_statevector()
state = sim.run(qcCX).result().get_statevector(qcCX)

#Get output
print(state.data)
```

This output means [0, 0, 0, 1] which means ∣11⟩. As the control qubit was '1', the target qubit was flipped.

## Toffoli Gate

Also called CCNOT gate

If both the control qubits are ∣1⟩, applies an X gate to the target.

If one or both the control qubits are ∣0⟩, does nothing

```python
#Create a circuit
qcCCX = QuantumCircuit(3)

#Make both the control qubits '1'
qcCCX.x(0)
qcCCX.x(1)

#Apply the CCNOT gate
qcCCX.ccx(0, 1, 2)

#Run on Aer simulator
sim = Aer.get_backend('aer_simulator')
qcCCX.save_statevector()
state = sim.run(qcCCX).result().get_statevector(qcCCX)

#Get output
print(state.data)
```

This output means [0, 0, 0, 0, 0, 0, 0, 1] which means ∣111⟩. As both the control qubits were '1', the target qubit was flipped.

## Phase Gate

Rotation by a specific angle about the z axis

```python
#Create a circuit
qcP = QuantumCircuit(1)

#Phase gate
qcP.h(0)
qcP.p(3.14/4,0)

#Run on Aer sim
sim = Aer.get_backend('aer_simulator')
qcP.save_statevector()
state = sim.run(qcP).result().get_statevector(qcP)

#Get state
print(state.data)
```
